How to iterate over rows in Pandas Dataframe

```python
import pandas as pd
inp=[{'Name':'Sameer','Age':25},
     {'Name':'Smriti','Age':21},
     {'Name':'Sujata','Age':27}]
df=pd.DataFrame(inp)
df
```

Using iterrows()

```python
for index,row in df.iterrows():
    print(row['Name'],row['Age'])
```

Using itertuples()

```python
for row in df.itertuples():
    print(getattr(row,'Name'))
```

Using index method

```python
for ind in df.index:
    print(df['Name'][ind],df['Age'][ind])
```

Using loc[] method

```python
for i in range(len(df)):
    print(df.loc[i,'Name'],"\t",df.loc[i,'Age'])
```

Using apply 

```python
print(df.apply(lambda row:row['Name']+ "  " +str(row["Age"]),axis=1))
```

Selecting rows in pandas DataFrame based on conditions

```python
record = {
 'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
 'Age': [21, 19, 20, 18, 17, 21],
 'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
 'Percentage': [88, 92, 95, 70, 65, 78] }
df=pd.DataFrame(record)
df
```

Insert row at given position in Pandas Dataframe

```python
df = pd.DataFrame({'Date':['10/2/2011', '12/2/2011', '13/2/2011', '14/2/2011'],
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy'],
                    'Cost':[10000, 5000, 15000, 2000]})
df
```

```python
def insert_row(rn,df,rv):
    df1=df[0:rn]
    df2=df[rn:]
    df1.loc[rn]=rv
    result=pd.concat([df1,df2])
    return result
```

```python
rn=1
rv={'Date':'11/2/2011','Event':'Act','Cost':17000}
if rn>df.index.max()+1:
    print("Invalid row number")
else:
    print(insert_row(rn,df,rv))
```

Create a list from rows in Pandas dataframe

```python
df = pd.DataFrame({'Date':['10/2/2011', '12/2/2011', '13/2/2011', '14/2/2011'],
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy'],
                    'Cost':[10000, 5000, 15000, 2000]})
row_list=[]
for index,rows in df.iterrows():
    curr=[rows['Date'],rows['Event'],rows['Cost']]
    row_list.append(curr)

print(len(row_list))
print(row_list[:2])
```

Ranking Rows of Pandas DataFrame

```python
student_details = {'Name':['Raj', 'Raj', 'Raj', 'Aravind', 'Aravind', 'Aravind',
                             'John', 'John', 'John', 'Arjun', 'Arjun', 'Arjun'],
                   'Subject':['Maths', 'Physics', 'Chemistry', 'Maths', 'Physics',
                             'Chemistry', 'Maths', 'Physics', 'Chemistry', 'Maths',
                                                           'Physics', 'Chemistry'],
                   'Marks':[80, 90, 75, 60, 40, 60, 80, 55, 100, 90, 75, 70]
               }
df=pd.DataFrame(student_details)
df['Ranking']=df['Marks'].rank(ascending=0)
df.set_index('Ranking',inplace=True)
df
```

```python
df.sort_index(ascending=True,inplace=True)
df
```

Convert a column to row name/index in Pandas

```python
data = {'name':["Akash", "Geeku", "Pankaj", "Sumitra", "Ramlal"],
       'Branch':["B.Tech", "MBA", "BCA", "B.Tech", "BCA"],
       'Score':["80", "90", "60", "30", "50"],
       'Result': ["Pass", "Pass", "Pass", "Fail", "Fail"]}
 
df = pd.DataFrame(data)
df1=df.pivot(index='Result',columns='name')
df1
```
